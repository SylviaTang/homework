```python
df_pop= pd.read_table("pop.csv",  
                       encoding = 'utf8',       
                       header = 3,              
                       skiprows = [35,36],      
                       sep =','                 
                      )
df_liferefuse= pd.read_table("life refuse.csv",   
                       encoding = 'utf8',       
                       header = 3,              
                       skiprows = [35,36],      
                       sep =','                 
                      )
```

```python
df1 = df_pop
```

```python
df1.columns
```

```python
df1.index
```

```python
df1['地区'].head()
```

```python
df1.columns = [int(x.replace('年','')) if x!='地区' else x for x in df1.columns]
df1.columns
```

```python
years = [x for x in df1.columns if x!='地区' ]
max(years)
```

```python
df1_melted = pd.melt(df1,id_vars=['地区'],value_vars=years)
```

```python
df1_melted.columns
```

```python
df1_melted.columns = ['地区','年','POP']
```

```python
df_pop_done = df1_melted
df_pop_done
```

```python
df2 = df_liferefuse
df2.columns = [int(x.replace('年','')) if x!='地区' else x for x in df2.columns]
years = [ x for x in df2.columns if x!='地区'] 
df2_melted = pd.melt(df2, id_vars=['地区'], value_vars=years)
df2_melted.columns = ['地区', '年', 'Liferefuse']
df_liferefuse_done = df2_melted
df_liferefuse_done
```

```python
df_p = df_pop_done.join(df_liferefuse_done,lsuffix='', rsuffix='_liferefuse')
df_p
```

```python
df_p.drop(["地区_liferefuse","年_liferefuse"],axis =1)
```

```python
df_p.drop(["地区_liferefuse","年_liferefuse"], axis=1).set_index(["地区","年"])
```

```python
df_p1 = df_p.drop(["地区_liferefuse","年_liferefuse"], axis=1).set_index(["地区","年"])
```

```python
df_p1.query("年==2015")
```

```python
%matplotlib inline
df_p1.query("年==2015").plot(kind='scatter', x='POP', y='Liferefuse', 
          title = 'POP与Liferefuse的关系：分省数据')
```
